```python
import pandas as pd
import re
#Reading the dataset into a variable
df=pd.read_csv("Crop_Database_Final_Output.csv")
df = df.rename( columns={"Unnamed: 0": "Index"})
#Verifying whether the dataset is read into the variable
display(df)
```

### Need for Data Cleaning:

There are several attributes in the data set which are polulated by textual data.

Textual data is not very convinient for data analysis

Textutal data needs to be condensed to categorical data for the purpose of data analysis.

We achieve coversion of textual data to categorical data by identifying keywords (words that help categorize the record/row)

Once key words are identified, they are retained and the rest of the textual data, which is noise in this case, is discarded.

```python
#Column'Annual, biennial, or perennial' is populated by textual data.
#We need to retain the the Keywords: Annual, Biennial, Perinnial in the all their forms and discard the noise. 
#We achieve this by using an iteration of a simple IF/ELSE sequence on the data
#We also need to set a count for the categories a particular record falls into, this will be used for analysis 

#define a column that will be the count for no. of categories a particular record falls under
df["Count of Annual, biennial, or perennial"] = 0
#iteration for identification of key words using an IF/ELSE sequence nested inside a FOR loop
for i in df['Index']:
    a=df.at[i,'Annual, biennial, or perennial']
    a=str(a)
#setting up an alternate variable, with default value NULL, to hold the result of our keyword search 
    a2=[]
#updating the alternate variable as and when a keyword is encountered in the data
    if 'Annual'in a or 'annual' in a:
        a2.append("Annual")
    if 'Biennial'in a or 'biennial' in a:
        a2.append("Biennial")
    if 'Perennial'in a or 'perennial' in a:
        a2.append("Perinnial") 
#Alternate variable remains a NULL list in case no keyword is encountered
    else:
        a2=a2
#updating the original dataset with the values of the alternate variable, which is of type LIST
#LIST is coverted to STRING data before being entered into the data set for analysis
df.at[i,'Annual, biennial, or perennial']=','.join(map(str,a2))
df.at[i,"Count of Annual, biennial, or perennial"] = len(a2)
```
